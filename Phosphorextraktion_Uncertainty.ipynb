```python
import numpy as np
import sympy as sp


b10 = 17500
c10 = 7000
d10 = 12000
b78 = 0.8758
e78 = 81618
c209 = 0.025
b512 = 663
c512 = 1099
b518 = 70700
c518 = 84567
b517 = 64.5
c517 = 47.6
c566 = 0.45
input_data = np.asarray([b10,c10,d10,b78,e78,c209,b512,c512,b518,c518,b517,c517,c566])

def zielwertsuche(b551,b552,c551,c552):
    x = sp.Symbol('x')
    f = (b551+b552+x*31/98)/(c551+c552+x) -0.166
    x = sp.solve(f)
    return x
def model1(c67,c322,c328):
    faktor = [c67,c322,c328]
    
    wk = 2.834
    ssk = 0.3
    psk = 0.55
    c23 = input_data[2]
    c25 = input_data[2] * wk
    c26 = input_data[2] * ssk
    c31 = input_data[2] * psk
    f78 = input_data[2] * input_data[4]/1000000
    RIP = input_data[3] * faktor[0]
    g78 = RIP * f78
    h78 = f78 * (1-RIP)
    c89 = c23+c25+c26+c31
    #suspension1
    b152 = 0.02*faktor[0]
    c157 = 1.3
    c158 = c89/c157
    c160 = c158 * b152
    c161 = 1.02
    c162 = c160*c161
    c225 = g78
    d225 = c225 * input_data[5]
    c226 = 2088
    d226 = c226 * input_data[5]
    c227 = c225+c226
    d227 = c227 * input_data[5]
    c241 = c89+c162
    #suspension2
    a326 = 0.62634
    b326 = 0.37365
    c326 = 0.40943
    c335 = 1.5
    c337 = c335*c23
    c387 = c241+c337
    c390 =c387*(1-faktor[1])
    c392 = c390*a326
    c394 = c390*b326
    c396 = c394*c326 
    c407 = c390
    c413 = c396
    c326 = 0.41
    c330 = c326*faktor[2]
    c411 = c413/c330
    c409 = c407-c411
    c415 = c411 * (1-c330)
    c431 = c25
    c432 = 0.95
    c433 = c162
    c434 = c432*c433
    c435 = c337
    c436 = c415
    c437 = 1-c330
    c438 = c437 * faktor[1] * (c337+c241)
    c439 = c431+c434+c435-c436-c438
    d518 = input_data[0] * input_data[8]/1000000
    e518 = input_data[1] * input_data[9]/1000000
    f518 = d518+e518
    zwischen1 = input_data[2]*input_data[4]/1000000*input_data[3]*faktor[0]
    zwischen2 = input_data[2]*input_data[4]/1000000*(1-input_data[3]*faktor[0])
    b538 = (input_data[2]*input_data[4]/1000000-zwischen1*input_data[5]-zwischen2)*(1-faktor[1])
    b539 = c31*31/98*(1-input_data[5])*(1-faktor[1])
    b540 =b538+b539
    b551 = b540
    b552 = f518
    c427 = c330*faktor[1] * (c337+c241)
    c551 = c23+c26+c31+0.05*c162-c413-c427
    c552 =input_data[0]+input_data[1]
    c553 = zielwertsuche(b551,b552,c551,c552)[0]
    c575 = c551+input_data[0]+input_data[1]+c553
    c576 = c575/input_data[-1]*(1-input_data[-1])
    c577 = c439
    c559 = c553/0.75*0.25
    c578 =c559
    c579 = c576-c577-c578
    
    e72 =656
    f72 = input_data[2]*e72/1000000
    g72 = f72*0.23362*faktor[0]
    h72 = f72*(1-0.23362*faktor[0])
    d220 = f72*0.23362*faktor[0]*0.99
    c295 = f72- f72*(1-0.23362*faktor[0])-f72*0.23362*faktor[0]*0.99
    b452 =c295*(1-faktor[1])
    b624 = b452
    f512 = (input_data[0]*input_data[6]+input_data[1]*input_data[7])/1000000
    c624 = f512
    e624 = b624+c624
    f624 = c575
    g624 =e624/f624*1000000
    b686 =g624
    
    e77 =125
    f77 = input_data[2]*e77/1000000
    g77 = f77*0.43577*faktor[0]
    h77 = f77*(1-0.43577*faktor[0])
    d224 = f77*0.43577*faktor[0]*0.025
    c300 = f77- f77*(1-0.43577*faktor[0])-f77*0.43577*faktor[0]*0.025
    b457 =c300*(1-faktor[1])
    b629 = b457
    f517 = (input_data[0]*input_data[-3]+input_data[1]*input_data[-2])/1000000
    c629 = f517
    e629 = b629+c629
    f629 = c575
    g629 =e629/f629*1000000
    b687 =g629
    
    e81 =2353
    f81 = input_data[2]*e81/1000000
    g81 = f81*0.43789*faktor[0]
    h81 = f81*(1-0.43789*faktor[0])
    d235 = f81*0.43789*faktor[0]*0.95
    c311 = f81- f81*(1-0.43789*faktor[0])-f81*0.43789*faktor[0]*0.95
    b468 = c311*(1-faktor[1])
    b633 = b468
    f512 = (input_data[0]*2260+input_data[1]*1989)/1000000
    c633 = f512
    e633 = b633+c633
    f633 = c575
    g633 = e633/f633*1000000
    b689 = g633
    
    c644 = c575
    c645 = 0.95
    c646 = c644/c645
    c647 = 0.05
    c648 = c646*c647
    c649 = c576-c648
    c655 = c31
    c662 = c553
    c664 = c655+c662
    c666 = c664/c646
    output = [c162, c411, c579, c664, c649, c646, c666, b686, b689]
    return output
```

```python
#  faktor = [c67,c322,c328]
model1(0.9,0.0128,1.0435)
```

```python
#定义抽样数量
sample_size = 30000

#定义faktor分布
f1 = np.reshape(np.random.uniform(low=0.66,high=0.9,size=sample_size),(sample_size,1))
f2 = np.reshape(np.random.triangular(0.0047,0.02,0.0242,size=sample_size),(sample_size,1))
f3 = np.reshape(np.random.uniform(low=1,high=1.082,size=sample_size),(sample_size,1))
faktor_sample=np.hstack((f1,f2,f3))

output_sample = []
for f in faktor_sample:
    output_sample.append(model1(f[0],f[1],f[2]))
```

```python
#可视化
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
output = []
for i in range(9):
    y = np.asarray(output_sample,dtype=float)[:,i]
    plt.hist(y, bins=40, facecolor="blue", edgecolor="black", alpha=0.7)
    plt.show()
    output.append(y)
np.shape(output)
```
